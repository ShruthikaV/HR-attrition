```python
# Calculate overall attrition rate
overall_attrition = (df['Attrition'] == 'Yes').mean() * 100

# Analyze attrition by demographics
attrition_by_gender = df.groupby(['Gender', 'Attrition']).size().unstack()
attrition_by_gender_pct = (attrition_by_gender.div(attrition_by_gender.sum(axis=1), axis=0) * 100)

attrition_by_education = df.groupby(['Education', 'Attrition']).size().unstack()
attrition_by_education_pct = (attrition_by_education.div(attrition_by_education.sum(axis=1), axis=0) * 100)


# Plotting
plt.figure(figsize=(15, 5))

# Plot 1: Gender
plt.subplot(151)
attrition_by_gender_pct['Yes'].plot(kind='bar')
plt.title('Attrition Rate by Gender')
plt.ylabel('Attrition Rate (%)')

# Plot 2: Education
plt.subplot(152)
attrition_by_education_pct['Yes'].plot(kind='bar')
plt.title('Attrition Rate by Education Level')
plt.ylabel('Attrition Rate (%)')


plt.tight_layout()
plt.show()

print(f"Overall Attrition Rate: {overall_attrition:.1f}%")
print("\Attrition Rate by Gender (%):")
print(attrition_by_gender_pct['Yes'])
print("\Attrition Rate by Education Level (%):")
print(attrition_by_education_pct['Yes'])
```
